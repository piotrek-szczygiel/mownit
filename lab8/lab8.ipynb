```julia
using FFTW, Plots
t1 = 0:1/1024:1
x1 = sin.(2*pi*t1*200) + 2*sin.(2*pi*t1*400)

plot(t1, x1)
```

```julia
y1 = fft(x1)

sticks(abs.(y1))
```

```julia
t2 = 0:1/1024:1
x2 = cos.(2*pi*t2*20)

display(plot(t2, x2))
display(sticks(abs.(fft(x2))))
```

```julia
x2_noise = zeros(size(x2, 1))

for i in 1:size(x2, 1)
    x2_noise[i] = x2[i] - 0.5 + rand()
end

y2_noise = fft(x2_noise)

display(plot(t2, x2_noise))
display(sticks(abs.(y2_noise)))
```

```julia
y2 = y2_noise

for i in 1:size(y2, 1)
    if abs(y2[i]) < 30
        y2[i] = 0
    end
end

sticks(abs.(y2))
```

```julia
x2_reduced = real.(ifft(y2))

display(plot(t2, x2_reduced))
```
